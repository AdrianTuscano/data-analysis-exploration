# Unethical nba players

This is going to use stats from the 25-26 nba season prior to the All-Star break to visualize the least ethical nba scorers based on points per game, free throws made, fouls drawn, attempted drives, and effective field goal percentage. This only covers the top 30 players in points per game. 

I am using my own made up stat, Free Thow Merchant Score(FTMS) which is a formula I made up, not based on any real statistics. <br>The formula is (Free throws made / (fouls drawn + (drives * 0.50))) * 100

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

dataframe = pd.read_csv('my_nba_player_data.csv')

x = dataframe['PPG']
y = ( dataframe['Free throws made'] / ((dataframe['Fouls Drawn']) + (dataframe['Drives'] * 0.5)) ) * 100 
s = (dataframe['eFG%'] - 0.45) * 10000
plt.figure(figsize=(8, 6))
plt.scatter(x, y, sizes = s, alpha=0.5)
plt.title('PPG vs Free Throw Merchant Score')

for idx, row in dataframe.iterrows():
    plt.annotate(row['Player'], (row['PPG'], y.iloc[idx]), 
                fontsize=8, ha='center')

plt.text(0.025, 0.95, 'Circle size represents eFG%', 
        transform=plt.gca().transAxes, fontsize=12,
        verticalalignment='top', bbox=dict(boxstyle='round', 
        facecolor='white', alpha=0.5))

plt.xlabel('PPG')       
plt.ylabel('Free Throw Merchant Score')
plt.grid()
plt.show()
```
