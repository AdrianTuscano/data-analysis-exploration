```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initializing data with some data on NBA players
data = pd.DataFrame({   
    'Name': ['Ant', 'Lebron', 'Curry', 'Durant', 'Booker'],
    'Ppg': [29, 22, 27, 26, 25],
    'City': ['New York', 'Los Angeles', 'San Francisco', 'Houston', 'Phoenix']
    })

# Using an object oriented approach to plot the data
fig, nba_ppg = plt.subplots()

nba_ppg.plot(data['Name'], data['Ppg'])
nba_ppg.set_xlabel('Player Name')
nba_ppg.set_ylabel('Points Per Game')
nba_ppg.set_title('Popular NBA Players PPG')
plt.show()
```

Line plots + multiple functions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initializing as an exponential function to plot. 
x = np.linspace(0.1, 10, 99)
y0 = np.sin(x)
y1 = np.cos(x)
y2 = -1 * np.sin(x)  
y3 = -1 * np.cos(x)

plt.plot(x, y0, label='sin(x)')
plt.plot(x, y1, label='cos(x)')
plt.plot(x, y2, label='-sin(x)')
plt.plot(x, y3, label='-cos(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric Wave Functions')
plt.grid(True)
plt.legend()
plt.show()
```
